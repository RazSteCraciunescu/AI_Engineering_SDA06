# Lucrul cu șiruri de caractere

Putem folosi anumiți operatori cunoscuți pentru a genera anumite funcționalități atunci când sunt aplicați unui șir de caractere.

````python
# Exemplu 1 -- len( )

print(len("Python"))

# Putem obține lungimea unui șir folosind funcția len( )
````

````python
# Exemplu 2 -- Accesarea, parcurgerea și întoarcerea unui șir

nume_utilizator = "Ana Popescu"
print(nume_utilizator[2])
print(nume_utilizator[-1])

# Putem accesa caractere de la capătul șirului.

print(nume_utilizator[0:3])   # De la poziția de START, până la poziția de STOP,
                              # EXCLUSIV!

print(nume_utilizator[4::]) # Toate caracterele de după poziția din stânga

print(nume_utilizator[::-1])  # Toate caracterele de înainte de poziția din
                              # dreapta, însă în sens opus.
````

````python
# Exemplu 1 -- „+"

print('Nume' + 'Prenume')

# OBS.: Operatorul „+" devine operator de CONCATENARE.
````

````python
# Exemplu 1 -- „*"

a = 'Python ' * 3
print(a)

# OBS.: Operatorul „*" devine operator de REPETIȚIE.
````

````python
# Exemplu 1 -- ' ' și " "

text = "It's a string"

text = 'It's not a string'

# OBS.: Nu putem folosi apostroful în mijlocul string-ului dacă definim
# string-ul cu ' '

my_text = 'It\'s also a string'

# Va trebui să anulăm funcționalitatea apostrofului prin '\'.
````

# Formatări de bază pentru șiruri de caractere

Formatarea șirurilor de caractere permite inserarea dinamică a unor valori într-un text, astfel încât acesta să fie ajustat în funcție de valori specifice.

În Python, există mai multe metode pentru formatarea șirurilor:

1.	Operatorul %

Operatorul % este una dintre metodele mai vechi de formatare a șirurilor în Python. Acesta funcționează similar cu printf din limbajul C.

```
print("text %tip_format" % valoare)
```

* %s - Șir de caractere (string)
* %d - Număr întreg (digit)
* %i - Număr întreg (integer)
* %f - Număr rațional (float)
* %c - Caracter ASCII (character)

2.	Metoda .format()

Metoda .format() oferă o alternativă mai flexibilă și mai modernă pentru operatorul %. Valorile pot fi inserate în șiruri folosind acolade {}.
```
print("Text {0} text {1}".format(valoare1, valoare2))
```


3.	f-strings (format string literals)

f-strings reprezintă metoda cea mai modernă și simplă pentru formatarea șirurilor de caractere în Python (introduse în Python 3.6).
Aceasta permite inserarea valorilor direct într-un șir folosind variabile în acolade.
```
print(f"Text {variabilă} text")
```



````python
# Exemplu 1

nume_utilizator = input("Cum te numești? ")
print("Salut %s!" % nume_utilizator)
````

````python
# Exemplu 2

print("Salut {0}!".format(nume_utilizator))

# OBS.: Numerotarea poate continua, dacă avem mai multe variabile.
````

````python
# Exemplu 3

print(f"Salut {nume_utilizator}!")
````

# Metode esențiale pentru verificarea și manipularea șirurilor

Tipul de date **string** în Python oferă o suită de metode prin care putem verifica anumite condiții sau manipula caracterele sale, în funcție de situația dată.

## Metode de verificare:

````python
# .isalnum( ) -- Verifică dacă un șir conține doar caractere alfanumerice
# (litere și cifre).

print("abc123".isalnum())   # True
print("abc 123".isalnum())  # False (spațiul nu este alfanumeric)
````

````python
# .isalpha( ) -- Verifică dacă toate caracterele sunt litere.

print("abc".isalpha())    # True
print("abc123".isalpha()) # False
````

````python
# .isdigit( ) -- Verifică dacă toate caracterele sunt cifre.

print("123".isdigit())    # True
print("Test10".isdigit()) # False
````

````python
# .islower( ) -- Verifică dacă toate literele sunt minuscule.

print("python".islower())  # True
print("Python".islower())  # False
````

````python
# .isupper( ) -- Verifică dacă toate literele sunt majuscule.

print("PYTHON".isupper())  # True
print("Python".isupper())  # False
````

````python
# .startswith( ) -- Verifică dacă șirul începe cu un anumit prefix.

print("Python".startswith("Py"))  # True
````

````python
# .endswith( ) -- Verifică dacă șirul se termină cu un anumit sufix.

print("Python".endswith("Py"))  # False
````

````python
# .find( ) -- Întoarce poziția primei apariții a subșirului căutat.

print("Ana are mere".find("mere"))  # 8
print("Ana are mere".find("test"))  # -1
````

## Metode de manipulare a șirului de caractere:

````python
# .lower( ) -- Transformă toate caracterele din șir în minuscule.

print("PYTHON".lower())  # python
````

````python
# .upper( ) -- Transformă toate caracterele din șir în majuscule.

print("python".upper())  # PYTHON
````

````python
# .strip( ) -- Elimină caractere de la începutul și sfârșitul șirului.

print("  Python  ".strip())  # "Python"

print(" .Python...".strip("."))  # " .Python"

# OBS.: Dacă nu includem niciun parametru, metoda va elimina automat spațiile
# albe.
````

````python
# .replace() -- Înlocuiește toate aparițiile unui subșir cu altul.

print("Ana are mere".replace("mere", "pere"))  # Ana are pere

print("Ana are mere".replace("m", "p")) # Ana are pere

# OBS.: Putem folosi și un singur caracter pentru a efectua operația.
````

````python
# .split() -- Împarte șirul într-o listă de subșiruri.

print("Ana are mere".split())  # ['Ana', 'are', 'mere']

print("Ana,Ion,Matei".split(",")) # ['Ana', 'Ion', 'Matei']

# OBS.: Dacă nu adăugăm niciun parametru, se vor împărți automat cuvintele după
# spațiile albe.
````

````python
# .join( ) -- Unește elementele unei liste într-un singur șir, folosind un
# separator specificat.

cuvinte = ["Ana", "are", "mere"]
print(" ".join(cuvinte))  # Ana are mere

# OBS.: Metoda .join( ) se apelează asupra separatorului care va apărea între
# elementele din listă.
````
