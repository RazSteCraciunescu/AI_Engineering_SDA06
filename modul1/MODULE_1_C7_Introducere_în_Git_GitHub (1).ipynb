## Capitolul 1: Introducere în Git

Ai pierdut vreodată paragrafe întregi dintr‑un document pentru că nu ai salvat la timp? În programare putem pierde funcționalități din aceeași cauză sau când două persoane modifică aceeași porțiune de cod. **Sistemele de control al versiunilor** precum Git rezolvă aceste probleme, permițând urmărirea și revenirea la orice versiune salvată.

Mai precis, Git este un sistem de control al versiunilor distribuit (VCS - Version Control System) care permite urmărirea modificărilor în fișiere și colaborarea eficientă între mai mulți dezvoltatori.

### De ce să folosim Git?

- Urmărirea istoriei modificărilor codului
- Colaborare facilă între programatori
- Posibilitatea de a reveni la o versiune anterioară
- Ramificare (branching) pentru lucrul pe mai multe funcționalități simultan
- Salvare în cloud cu GitHub, GitLab sau Bitbucket

### Ce este un depozit (repository)?

Un **repository** este un dosar care conține fișierele proiectului *și* istoricul lor complet. El poate fi:

- **local** – pe calculatorul tău, accesibil fără internet;
- **remote** – pe un server (GitHub, GitLab, HuggingFace), unde mai mulți dezvoltatori colaborează.

## Capitolul 2: Lucrul cu Git Local

### Stările unui fișier în Git

1. **Untracked** – fișier nou, neverificat
2. **Staged** – fișier adăugat pentru următorul commit
3. **Committed** – modificare salvată în istorie

Odată ce noi inițiem un nou repository, toate fișierele deja existene și orice fișier nou adăugat sunt în starea de **untracked**.

Va trebui să adăugăm constant fișierele noi sau modificările aduse la fișierele deja existente în repository-ul nostru, pentru a asigura că ele nu se pierd.

Git-Architechture-Git-Tutorial-Edureka-2.png

### __EXERCIȚIU__ Inițierea unui repository local.

1. Creează un nou director.

2. Deschide fișierul cu VSCode/PyCharm și creează un fișier .py în interiorul său. Scrie ceva în acel fișier.

3. Din directorul nou, copiază adresa sa. (Click dreapta pe director din navigatorul din File Explorer.)

4. Deschide o nouă consolă și rulează comanda:
```
cd adresa/pana/la/directorul/tau
```

5. Rulează comanda următoare pentru a iniția un repository local:
```
git init
```

6. Pregătește versiunea actuală a tuturor fișierelor din directorul tău pentru a fi adăugate în repository-ul local folosind:
```
git add .
```
La acest pas, spunem că fișierele sunt **staged**.

7. Salvează modificările în repository-ul local cu un mesaj descriptiv folosind:
```
git commit -m "mesaj informativ pentru fisierele adaugate"
```

## Capitolul 3: Lucrul cu Git Remote

Pentru a garanta faptul că fișierele noastre sunt salvate orice s-ar întâmpla cu directorul nostru, putem crea o copie a acestui repository local pe care l-am creeat mai devreme pe GitHub.



### __EXERCIȚIU__ Crearea unui repository pe GitHub

1. Mergi pe https://github.com.
2. Creează un nou repository pe contul tău, în secțiunea *Repositories*.

! Important: Asigurați-vă că nu adăugați niciun fișier `README.md` și că lăsați repository-ul public.
3. Copiază linkul cu HTTPS sau SSH.

### Conectarea celor două repository-uri

Acum, ultimul pas care mai rămâne de făcut este conectarea repository-ului public cu cel de la noi din calculator. Acest lucru se poate face ușor folosind un `remote`. Remote-urile sunt modul prin care Git știe cu ce repository de pe GitHub să comunice.


De aceea, avem nevoie de acel link pe care l-ați copiat mai devreme.

### __EXERCIȚIU__ Conectarea cu GitHub

1. În consola deschisă mai devreme, rulează comanda următoare pentru a stabili conexiunea:
```bash
git remote add origin https://github.com/utilizator/nume-repo.git
```
! Schimbă link-ul dat cu cel copiat mai devreme.

2. Adaugă toate schimbările locale pe care le-ai făcut în repository-ul local pe cel găzduit de GitHub folosind următoarea comandă:
```bash
git push -u origin master
```

*Felicitări!* Dacă totul a mers cum trebuie, ar trebui să poți da Refresh la pagina de GitHub asociată repository-ului tău pentru a vedea schimbările.

## Capitolul 4: Ramuri diferite de dezvoltare

Git ne permite să facem ceva numit *version control*. Acest proces implică și verificarea sau testarea de diferite versiuni ale codului nostru, într-un mod organizat și centralizat.
Vom putea realiza acest lucru folosind conceptul de *branches* sau ramuri, pe care le vom comasa numai atunci când garantăm performanța acestora prin testări și verificări ale codului.



### __EXERCIȚIU__ Crearea și unirea unui branch cu cel principal

1. Folosește comanda următoare pentru a crea un nou branch la proiectul tău (o nouă versiune pornind de la starea actuală a sistemului):
```bash
git checkout -b nume_ramura
```
Această comanda creează o nouă ramură cu numele dat și o și accesează. Dacă dorim să facem acest proces în doi pași diferiți vom rula:
```bash
git branch nume_ramura
git checkout nume_ramura
```

2. Modifică ceva în fișierul .py din proiectul tău.

3. Adu modificările înapoi în branch-ul principal folosind:
```bash
git merge nume_ramura
```
