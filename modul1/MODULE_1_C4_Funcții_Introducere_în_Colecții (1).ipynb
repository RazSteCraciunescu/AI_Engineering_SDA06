# Funcții în Python

## Ce este o funcție?

O funcție este un bloc de cod reutilizabil care execută o anumită sarcină. Funcțiile ne permit să scriem mai puțin cod și să-l reutilizăm ori de câte ori este nevoie. Ele pot accepta parametri (date de intrare) și pot returna o valoare.

##De ce să folosim funcții?
	1.	Reutilizare – Scriem o funcție o dată și o putem folosi ori de câte ori dorim.
	2.	Claritate – Codul devine mai organizat și mai ușor de citit.
	3.	Întreținere – Dacă trebuie să schimbăm logica, o facem într-un singur loc.
	4.	Modularitate – Putem împărți un program complex în funcții mai mici și mai simple.

## Sintaxă
```
def numele_functiei(parametri):
    # Blocul de cod care se execută atunci când funcția este apelată
    return valoare
```
* def -	Cuvânt cheie care definește funcția
* numele_functiei -	Numele funcției (fără spații, de preferat în snake_case)
parametri	Variabilele care primesc datele de intrare
* return - Returnează valoarea la sfârșitul funcției (opțional)
* blocul de cod -	Instrucțiunile care se execută atunci când funcția este apelată

````python
# EXEMPLU 1: Funcție fără parametri

def salut():
    print("Funcția a fost apelată. Salut!")

# Apelarea funcției
salut()

print(type(salut()))  # Dacă inserăm paranteze, înseamnă că apelăm funcția și
                      # lucrăm cu rezultatul ei.

print(type(salut))    # Dacă nu includem parantezele, lucrăm cu definiția în
                      # sine a funcției, deci îi putem vedea tipul de date
                      # special „function".

print(salut)          # Dacă printăm funcția în sine, putem vedea adresa din
                      # memorie unde este salvată.

print(bool(salut))    # Valoarea de adevăr a unei funcții în sine e True.

print(bool(salut()))  # Valoarea de adevăr a „NoneType" (tipul de date obținut
                      # dacă nu întoarcem nimic) este False

#OBS.: Putem defini o funcție fără nicio valoare întoasă (fără niciun return).

# Putem apela o funcție oricând, însă de fiecare dată când lucrăm cu rezultatul
# ei (adică atunci când punem parantele pentru parametri), funcția va rula
# complet (în exemplul de față, chiar dacă am vrut să obținem doar tipul de
# date al rezultatului funcției la primul print( ), funcția a trecut oricum prin
# tot procesul de a scrie pe ecran mesajul dorit.)
````

# Parametri Impliciți în Funcții (Default Arguments)

Un parametru implicit este un parametru al unei funcții care are o valoare prestabilită (default). Aceasta înseamnă că, dacă nu specificăm acel parametru atunci când apelăm funcția, Python folosește valoarea implicită.

## De ce folosim parametrii implicați?
	1.	Flexibilitate – Putem apela funcția fără a specifica toți parametrii.
	2.	Ușurință în utilizare – Facem apelurile funcțiilor mai clare și mai simple.
	3.	Valori implicite logice – Când nu oferim o valoare specifică, putem folosi una “de bază”.

### Sintaxă
  ```
  def numele_functiei(parametru1, parametru2=valoare_implicita):
    # Codul funcției
  ```
* Parametrii fără valoare implicită trebuie să fie poziționați înaintea celor cu valoare implicită.
* Nu putem pune un parametru cu valoare implicită urmat de un parametru fără valoare implicită.

# COLECȚII

Colecțiile sunt structuri de date care ne permit să stocăm și să organizăm mai multe valori într-un singur obiect. Acestea sunt esențiale pentru programarea eficientă, deoarece ne ajută să lucrăm cu grupuri de date (de exemplu, liste de nume, mulțimi de produse, perechi cheie-valoare etc.).

## De ce sunt importante colecțiile?
	•	Organizarea datelor – Putem stoca și gestiona mai multe valori într-un singur obiect.
	•	Acces și manipulare eficientă – Accesarea, adăugarea, ștergerea și modificarea elementelor se face ușor.
	•	Flexibilitate – Fiecare tip de colecție are scopul său (liste ordonate, seturi unice, dicționare cu perechi cheie-valoare).
	•	Reutilizare – Folosim colecțiile pentru stocarea datelor în mod eficient, mai ales când lucrăm cu bucle și funcții.

## Tipuri de colecții în Python
În Python, există 5 tipuri principale de colecții: liste (list), tupluri (tuple), mulțimi (set), dicționare (dict) și șiruri de caractere (string). Fiecare dintre aceste colecții are caracteristici unice, care le fac utile în contexte diferite. Mai jos, vom explica fiecare tip de colecție în propoziții clare și ușor de înțeles.

image.png


### Liste (list)
O listă este o colecție **ordonată** și **mutabilă**, ceea ce înseamnă că putem modifica elementele din listă după ce au fost create.
Listele permit accesarea elementelor folosind **indexul** și putem adăuga, șterge sau modifica elementele existente.
Este permisă **duplicarea** elementelor, adică aceeași valoare poate apărea de mai multe ori într-o listă.

#### Sintaxă
```
lista = ['element1', 'element2', 12, 'element1', True]
```

### Tupluri (tuple)
Un tuplu este o colecție **ordonată**, dar **imutabilă**, ceea ce înseamnă că nu putem modifica elementele după ce tuplul a fost creat.
Similar listelor, putem accesa elementele prin **index**, **dar nu putem adăuga, șterge sau modifica** elementele.
Tuplurile permit, de asemenea, **duplicarea** elementelor, ceea ce înseamnă că aceeași valoare poate apărea de mai multe ori într-un tuplu.

#### Sintaxă
```
tuplu = ('element1', 'element2', 12, 'element1', True)
```

### Seturi (set)
O mulțime (un set) este o colecție neordonată de elemente unice, ceea ce înseamnă că nu sunt permise duplicatele.
Fiind neordonată, nu putem accesa elementele folosind un index, dar putem itera prin ele.
Seturile sunt mutabile, așa că putem adăuga și elimina elemente.

#### Sintaxă
```
numere = {1, 2, 3, 4, 5}
```

### Dicționare (dict)
Un dicționar este o colecție de perechi **cheie-valoare**.
Fiecare **cheie** trebuie să fie **unică**, iar valorile sunt asociate acestor chei.
Putem accesa valorile prin cheia corespunzătoare, iar ordinea elementelor este **păstrată** (de la Python 3.7 încolo).
Dicționarele sunt **mutabile**, ceea ce înseamnă că putem adăuga, șterge sau modifica perechile cheie-valoare.

#### Sintaxă
```
student = {'nume': 'Andrei', 'varsta': 25, 'oras': 'București'}
```
! Cheia este elementul din membrul stâng al „:” și valoarea este cel din membrul drept.

### Șiruri de caractere (String)
Un șir de caractere (string) este o secvență de **caractere** (litere, simboluri sau spații).
Un șir de caractere este **ordonat**, ceea ce înseamnă că putem accesa fiecare caracter prin **index**.
Șirurile sunt **imutabile**, ceea ce înseamnă că **nu putem modifica caracterele individuale**, dar putem crea un nou șir de caractere.

#### Sintaxă
```
eleva = "Maria"
```
