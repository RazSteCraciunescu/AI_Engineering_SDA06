#Capitol 4: Type Conversion (Conversia Tipurilor de Date)


Conversia tipurilor de date (Type Conversion) reprezintă procesul de transformare a unei valori dintr-un tip de date în altul. Acest proces este esențial în Python, mai ales atunci când lucrăm cu date de la utilizatori sau cu informații preluate din fișiere, unde datele pot fi stocate sub formă de text (string), dar trebuie folosite ca numere.

##Tipuri principale de conversii:
Conversii implicite – Realizate automat de Python atunci când tipul de date mai mic este transformat într-un tip mai “cuprinzător” (de exemplu, int la float).

Conversii explicite – Realizate manual de programator, folosind funcții de conversie.

### Conversii implicite

Python realizează conversii implicite în situații simple. De exemplu, dacă adunăm un int cu un float, Python va converti automat int la float, pentru a evita pierderea de date.

```python
# Exemplu

numar_int = 5
numar_float = 3.2

rezultat = numar_int + numar_float  # Python convertește automat 5 la 5.0

print("Rezultatul este:", rezultat)  # Va afișa 8.2
print("Tipul rezultatului este:", type(rezultat))  # Va afișa <class 'float'>

# OBS.: Putem folosi funcția type( ) pentru a determina tipul de date al unei
# variabile.
```

### Conversii explicite (Type Casting)

Conversiile explicite sunt realizate prin utilizarea funcțiilor de conversie. Cele mai folosite funcții de conversie sunt:

**int(x)** – Converteste x la număr întreg.

**float(x)** – Converteste x la număr zecimal.

**str(x)** – Converteste x la șir de caractere (string).

**bool(x)** – Converteste x la valoare booleană (True sau False).


```python
# Exemplu 1 -- int > float

numar_text = "1"
numar_int = int(numar_text)
numar_float = float(numar_text)

print("Număr întreg:", numar_int)  # Afișează 10
print("Număr rațional:", numar_float)  # Afișează 10.0
```

```python
# Exemplu 2 -- int > string | float > string

numar_int = 42
numar_float = 3.14159

text_int = str(numar_int)
text_float = str(numar_float)

print("Număr întreg convertit la string:", text_int)  # Afișează '42'
print(type(text_int))

print("Număr rațional convertit la string:", text_float)  # Afișează '3.14159'
print(type(text_float))
```

```python
# Exemplu 3 -- !! Conversia in BOOLEAN

# ORICE VALOARE IN PYTHON ARE SI O VALOARE DE ADEVAR ASOCIATA.

print(bool(0))  # False
print(bool(1))  # True
print(bool(-10))

print()

print(bool(""))  # False
print(bool(" "))
print(bool("Python"))  # True

print()

print(bool(0.0)) # False
print(bool(0.6)) # True
```

# Capitol 5: Operații fundamentale și operații speciale specifice Python. OPERATORI

Operațiile fundamentale în Python includ operații aritmetice, de atribuire, logice și de comparație. Pe lângă acestea, Python are și câteva operații speciale care nu se regăsesc în toate limbajele de programare, cum ar fi „împărțirea întreagă” sau „ridicarea la putere” listelor.

```python
import pandas as pd

# Definirea datelor pentru tabel
data = {
    'Operator': ['+', '-', '*', '/', '//', '%', '**'],
    'Descriere': ['Adunare', 'Scădere', 'Înmulțire', 'Împărțire', 'Împărțire întreagă (Câtul)', 'Restul împărțirii', 'Ridicare la putere'],
    'Exemplu': ['2 + 3 = 5', '7 - 2 = 5', '3 * 4 = 12', '10 / 2 = 5.0', '10 // 3 = 3', '10 % 3 = 1', '2 ** 3 = 8']
}

# Crearea DataFrame-ului
df = pd.DataFrame(data)

# Afișarea tabelului

print(df)

# OBS.: Aici am evidențiat folosirea bibliotecii pandas, una din cele mai
# puternice biblioteci de Data Science. Vom reveni la pandas în cursurile
# viitoare. Până atunci, e important să înțelegeți conținutul tabelului.
```

```python
# Exemplu pentru operații aritmetice

a = 10
b = 3

# Operații aritmetice
print("Adunare:", a + b)
print("Împărțire întreagă:", a // b)
print("Restul împărțirii:", a % b)
print("Puterea:", a ** b)
```

! Python oferă o multitudine de operatori, unii dintre care nu i-am evidențiat aici. Puteți citi mai multe [aici](https://www.w3schools.com/python/python_operators.asp).

```python
# __EXERCIȚIU__ Se dau trei numere:
x = 10
y = 30
z = 5

# Să se calculeze:

# 1. Restul împărțirii lui x la y
print(x % y)

# 2. Rezultatul ridicării lui z la pătrat
print(z * z)
print(z ** 2)

# 3. Rezultatul înmulțirii celor trei numere
print(x * y * z)

# 4. Valoarea lui t = y / ((x + 2) * 5)
t = y / ((x + 2) * 5)
print(t)
```

## Operații logice. OPERATORI LOGICI

Aceste operații sunt folosite pentru a construi expresii logice (expresii care funcționează strict pe valori de adevăr -- True / False).

```python
# Definirea datelor pentru tabel
data = {
    'Operator': ['and', 'or', 'not'],
    'Descriere': ['Adevărat dacă ambele condiții sunt adevărate', 'Adevărat dacă cel puțin o condiție e adevărată', 'Inversează valoarea booleană'],
    'Exemplu': ['True and False = False', 'True or False = True', 'not True = False']
}

# Crearea DataFrame-ului
df = pd.DataFrame(data)

# Afișarea tabelului

print(df)
```

```python
# Exemplu pentru operații logice

x = True
y = False

print("x AND y:", x and y)  # False
print("x OR y:", x or y)  # True
print("NOT x", not x) # False
```
