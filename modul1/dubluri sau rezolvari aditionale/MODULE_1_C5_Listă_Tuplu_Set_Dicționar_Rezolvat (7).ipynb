# Liste
O listă este o colecție **ordonată** și **mutabilă** care poate conține diferite tipuri de date (numere, șiruri de caractere, dicționare, seturi, chiar și alte liste).
Listele sunt una dintre cele mai folosite structuri de date din Python datorită flexibilității și ușurinței în utilizare.

##Caracteristici principale ale listelor
	1.	Ordonată – Fiecare element din listă are un index. Ordinea elementelor este păstrată.
	2.	Mutabilă – Putem adăuga, șterge și modifica elementele din listă.
	3.	Permite duplicate – Listele permit elemente duplicate, adică același element poate apărea de mai multe ori.
	4.	Acceptă tipuri de date diferite – O listă poate conține valori de tipuri diferite (int, float, str, dict, list, etc.).

````python
# EXEMPLU 1
fructe = ['mere', 'pere', 'cireșe', 'mere']

print(fructe[0])  # Accesăm primul element ('mere')
print(fructe[2])  # Accesăm al treilea element ('cireșe')

# OBS.: Folosim aceeași notație "[ ]" ca atunci când dorim să accesăm
# caracterele din interiorul unui string.

# La fel, se păstrează și numerotarea elementelor începând de la 0.
````

````python
# __EXERCIȚIU__ Se dă lista urmatoare:
numere = [1, 5, 12, 43, 2]

# Sa se acceseze primul element din lista folosind pozitia sa.
print(numere[0])

# Sa se afiseze pe ecran numarul `43` prin accesarea pozitiei sale din lista.
print(numere[3])
````

## Adăugarea elementelor într-o listă

Avem trei variante de bază de a adăuga elemente într-o listă predefinită:
1. .append(x) -- Adaugă elementul x la finalul listei
2. .insert(i, x) -- Inserează elementul x pe poziția i și mută celelalte elemente mai la dreapta
3. .extend(lista2) -- Adaugă elementele dintr-o altă listă la finalul listei inițiale

!! TOATE ACESTEA SUNT METODE. În alte cuvinte, pentru a le putea rula, trebuie să le atașăm unei variabile care conține o listă.

````python
# EXEMPLU 1
fructe = ['mere', 'pere']
fructe.append('cireșe')  # Adaugă 'cireșe' la final
print(fructe)  # ['mere', 'pere', 'cireșe']

fructe.insert(1, 'banane')  # Inserează 'banane' pe poziția 1
print(fructe)  # ['mere', 'banane', 'pere', 'cireșe']

fructe.extend(['portocale', 'struguri'])  # Adaugă mai multe elemente
print(fructe)  # ['mere', 'banane', 'pere', 'cireșe', 'portocale', 'struguri']
````

## Ștergerea elementelor dintr-o listă

Avem patru variante de bază de a șterge elemente dintr-o listă predefinită:
1. .remove(x) -- Elimină prima apariție a elementului x
2. .pop(i) -- Elimină ȘI RETURNEAZĂ elementul de pe poziția i. De menționat că putem folosi folosi pop() și fără parametri pentru a elimina și returna ultimul element din listă.
3. del lista[i] -- Șterge elementul de pe poziția i. Aici reținem că putem folosi și „del lista” pentru a șterge lista cu totul
4. .clear() -- Șterge toate elementele listei, DAR NU ȘI LISTA ÎN SINE


````python
# EXEMPLU 1

fructe = ['mere', 'pere', 'cireșe', 'pere']
fructe.remove('pere')  # Elimină prima apariție a 'pere'
print(fructe)  # ['mere', 'cireșe', 'pere']

fruct = fructe.pop(1)  # Elimină și returnează elementul de la index 1
print(fructe)  # ['mere', 'pere']
print(fruct)

del fructe[0]  # Șterge elementul de la index 0
print(fructe)  # ['pere']

fructe.clear()  # Golește lista
print(fructe)  # []

del fructe
# print(fructe) # VA GENERA O EROARE DEOARECE fructe NU MAI ESTE DEFINIT
````

## Alte operații utile cu liste
* len(lista) -- Returnează lungimea listei
* lista.sort() -- Sortează lista crescător
* lista.reverse() -- Inversează ordinea elementelor
* min(lista) -- Returnează cel mai mic element din listă
* max(lista) -- Returnează cel mai mare element din listă
* sum(lista) -- Returnează suma elementelor

````python
# EXEMPLU 1

numere = [4, 1, 8, 3, 9]
print(len(numere))  # 5 (numărul de elemente)

numere.sort()  # Sortează lista în ordine crescătoare
print(numere)  # [1, 3, 4, 8, 9]

numere.reverse()  # Inversează ordinea
print(numere)  # [9, 8, 4, 3, 1]

print(min(numere))  # 1
print(max(numere))  # 9
print(sum(numere))  # 25
````

# Indexare. Slicing

Slicing-ul este o tehnică de a extrage o sub-listă dintr-o listă mai mare. În loc să accesăm doar un singur element (folosind un index), putem folosi intervale (range) de indecși pentru a obține mai multe elemente simultan.

Pentru a selecta o sub-listă dintr-o listă, folosim operatorul de slicing:
```
numele_listei[start:stop:pas]
```
	•	start – Indexul de unde începe selecția (inclusiv)
	•	stop – Indexul unde se oprește selecția (fără a include acest index)
	•	pas – Cât de mari să fie pașii între elementele selectate (implicit este 1)

Dacă start sau stop sunt omise, Python presupune că start = 0 și stop = len(lista).



````python
#  Exemplu 1
fructe = ['mere', 'pere', 'cireșe', 'banane', 'portocale', 'struguri']

sub_lista = fructe[1:4]
print(sub_lista)

# start = 1 (începem de la indexul 1, care este 'pere')
# stop = 4 (ne oprim înainte de indexul 4, deci ‘portocale’ nu este inclus)
# pas = 1 (selectăm fiecare element consecutiv)
````

````python
# __EXERCIȚIU__ Se dă lista urmatoare:
numere = [345, 123, 12, 364, 85, 5]

# Sa se ordoneze lista descrescator si sa se creeze o noua lista cu primele
# 3 elemente din lista sortata. Sa se afiseze noua lista.
````

````python
# Exemplu 2

fructe = ['mere', 'pere', 'cireșe', 'banane', 'portocale', 'struguri']
sub_lista = fructe[1:6:2]
print(sub_lista)
````

# Operatorul IN

Operatorul in este folosit pentru a verifica dacă un element se află într-o listă. Cu alte cuvinte, ne răspunde la întrebarea:

	“Acest element se găsește în listă?”

Operatorul in returnează o valoare de tip True (adevărat) sau False (fals), în funcție de prezența elementului în listă.

## Sintaxa:
```
element in lista
```
	•	element – Elementul pe care dorim să verificăm dacă se află în listă
	•	lista – Lista în care căutăm elementul
  


````python
# Exemplu 1

fructe = ['mere', 'pere', 'cireșe', 'banane']

# Verificăm dacă un element se află în listă
print('pere' in fructe)  # True (deoarece 'pere' se află în listă)
print('struguri' in fructe)  # False (deoarece 'struguri' nu este în listă)
````

````python
# Exemplu 2

fructe = ['mere', 'pere', 'cireșe', 'banane']

print(fructe.index('pere'))
print(fructe[1])

# OBS.: Putem afla orice poziție a unui element în listă folosind metoda
# .index( )
````

# Tupluri (Tuples) – Operații și Metode Asociate

Un tuplu este o colecție ordonată și imutabilă.
Spre deosebire de liste, tuplurile nu pot fi modificate după ce au fost create. Aceasta le face utile atunci când dorim să stocăm valori constante care nu ar trebui modificate.

## Caracteristici principale ale tuplurilor
	1.	Ordonate – Elementele au o poziție (index) și sunt păstrate în ordinea adăugării.
	2.	Imutabile – Nu putem schimba, adăuga sau șterge elemente după ce tuplul a fost creat.
	3.	Permite duplicate – Un tuplu poate conține elemente duplicate.
	4.	Acceptă tipuri de date diferite – Un tuplu poate conține valori de diferite tipuri (int, str, list, tuple, etc.).

  ## Sintaxă
```
numele_tuplului = (element1, element2, element3, ...)
```




````python
# Exemplu 1

tuplu = (5, 10)
tuplu_unic = (42,)  # Tuplu cu un singur element

print(tuplu, tuplu_unic)

# OBS.: Pentru a iniția un tuplu cu un singur element, trebuie să adăugăm
# virgulă la final

tuplu_unic_test = (42)
print(type(tuplu_unic_test))

# Dacă nu adăugăm virgula, interpretorul va presupune că dorim să asociem
# valoarea introdusă în paranteze (în acest caz un număr întreg -- int).
````

# Operații cu tupluri

Toate operațiile posibile în lucrul cu tupluri sunt acelea care se folosesc de tuplul în sine, NU DE ELEMENTELE LUI. În cazul listelor, puteam spre exemplu să adăugăm un singur element în listă, dar la tupluri vom vedea că putem adăuga doar un tuplu întreg în alt tuplu.

````python
# Exemplu 1

tuplu1  = (1,2,3,4)
tuplu2 = (5,6,7,8)

tuplu3 = tuplu1 + tuplu2 # Operația de concatenare se folosește prin operatorul „+"
print(tuplu3)

tuplu4 = tuplu1 * 4 # Putem repeta același tuplu prin folosirea operatorului „*"
print(tuplu4)

print(tuplu1[2]) # Putem doar să ACCESĂM elemente din tuplu, nu să le manipulăm.

print(2 in tuplu1)  # Operatorul IN funcționează deoarece nu schimbă valorile,
                    # doar le verifică.

print(len(tuplu1)) # Putem folosi funcția len( ) și în lucrul cu tupluri.
````

````python
# Exemplu 2: Metode asociate

tuplu = (1, 2, 3, 2, 4, 2)

# Count - de câte ori apare 2
numar_aparitii = tuplu.count(2)
print(f"Numărul de apariții ale lui 2: {numar_aparitii}")  # 3

# Index - poziția primei apariții a lui 3
pozitia = tuplu.index(3)
print(f"Prima poziție a lui 3: {pozitia}")  # 2
````

# Seturi (Set) – Operații și Metode Asociate

Un set este o colecție neordonată de elemente unice.
În contrast cu listele, seturile nu permit elemente duplicate, iar ordinea elementelor nu este garantată.
Un set este similar cu o punguță de obiecte: dacă bagi mâna, nu știi ce element vei extrage, dar știi că nu există duplicate.

## Caracteristici principale ale seturilor
	1.	Neordonate – Nu putem controla ordinea în care sunt stocate elementele.
	2.	Fără duplicate – Fiecare element dintr-un set este unic (dacă adăugăm un element deja existent, nu va fi adăugat din nou).
	3.	Mutabile – Putem adăuga și șterge elemente, dar seturile nu suportă modificarea directă a elementelor.
	4.	Imutabilitate a elementelor – Elementele unui set trebuie să fie valori imutabile (de exemplu, numere, șiruri de caractere, tupluri).


### Sintaxă
```
numele_setului = {element1, element2, element3, ...}
```
	1.	Folosind acolade {} – Aceasta este cea mai frecventă metodă.
	2.	Folosind funcția set() – Utilă pentru a converti liste sau alte colecții în seturi.


````python
# Exemplu 1

# Set cu mai multe elemente
fructe = {'mere', 'pere', 'cireșe'}

# Set cu elemente de tipuri diferite
mix = {1, 'Python', 3.14, True}

# Crearea unui set dintr-o listă (elimină duplicatele automat)
lista_numere = [1, 2, 2, 3, 4, 4, 5]
numere = set(lista_numere)
print(numere)  # {1, 2, 3, 4, 5}

# Set gol (ATENȚIE: {} este un dicționar gol, nu un set)
set_gol = set()
print(set_gol)  # set()

# OBS.: La ce trecerea de la o listă la un set, duplicatele sunt eliminate
# automat.
````

````python
# Exemplu 1

# Adăugare de elemente
fructe = {'mere', 'pere'}
fructe.add('cireșe')
print(fructe)  # {'mere', 'pere', 'cireșe'}

# OBS.: La seturi nu mai putem folosi metoda .append( ), ci apelăm la .add( )

# Uniune
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = set1.union(set2) # Echivalentul reuniunii din matematică
# set1 U set2
print(set3)  # {1, 2, 3, 4, 5}


# Intersecție
set4 = set1.intersection(set2) # Echivalentul intersecției din matematică
print(set4)  # {3}

# Diferență
set5 = set1.difference(set2) # Echivalentul diferenței de mulțimi
print(set5)  # {1, 2}

# Eliminarea unui element
fructe.remove('pere')
print(fructe)  # {'mere', 'cireșe'}

# Golirea unui set
fructe.clear()
print(fructe)  # set()
````

# Dicționare (Dictionary) – Operații și Metode Asociate

Un dicționar este o colecție de perechi cheie-valoare.
Este similar unui carnet de telefon sau unei enciclopedii, unde:

	*	Cheia (key) este cuvântul căutat (numele abonatului sau termenul enciclopedic).
	*	Valoarea (value) este informația asociată (numărul de telefon sau definiția termenului).

Dicționarele sunt folosite atunci când dorim să asociem valori cu chei și să accesăm valorile pe baza cheilor, nu a indexului.

### Sintaxă
```
numele_dicționarului = {
    cheie1: valoare1,
    cheie2: valoare2,
    cheie3: valoare3,
}
```

Metode de creare a unui dicționar:

	1.	Folosind acolade {} – Metoda preferată și mai intuitivă.
	2.	Folosind funcția dict() – Utilă pentru a crea dicționare mai dinamice.



````python
# Exemplu 1

# Dicționar de produse și prețuri
produse = {'mere': 5, 'pere': 7, 'cireșe': 15}
print(produse)
print(produse['pere'])

# Dicționar cu chei de tip string și valori mixte
student = {'nume': 'Andrei', 'varsta': 20, 'note': [8, 9, 10], 30:'colegi'}
print(student)

# OBS.: Într-un dicționar putem avea orice tip de date, atât pentru chei, cât
# și pentru valori, însă acest lucru este foarte DESCURAJAT. Aveți grijă când
# definiți un dicționar, întrucât cheia va fi mereu în stânga „:" și valoarea
# va fi în dreapta.

# Crearea unui dicționar gol
dict_gol1 = {}
dict_gol2 = dict()
````

# Accesarea elementelor dintr-un dicționar

Putem accesa o valoare dintr-un dicționar folosind cheia.
Dacă cheia nu există, obținem o eroare KeyError, dar putem folosi metoda .get() pentru a evita această eroare.


````python
# Exemplu 1

produse = {'mere': 5, 'pere': 7, 'cireșe': 15}

# Accesarea valorilor folosind cheia
print(produse['pere'])  # 7

# Utilizarea metodei .get()
valoare = produse.get('struguri', 'nu există')
print(valoare)  # 'nu există'

# Al doilea parametru al metodei .get( ) este valoarea IMPLICITĂ sau, în alte
# cuvinte, ce se întâmplă dacă nu găsește cheia căutată.
````

````python
# Exemplu 3: Metode de folosire a componentelor unui dicționar

produse = {'mere': 5, 'pere': 7, 'cireșe': 15}

# Obținerea tuturor cheilor
chei = produse.keys()
print(chei)  # dict_keys(['mere', 'pere', 'cireșe'])

# Obținerea tuturor valorilor
valori = produse.values()
print(valori)  # dict_values([5, 7, 15])

# Obținerea tuturor perechilor cheie-valoare
perechi = produse.items()
print(perechi)  # dict_items([('mere', 5), ('pere', 7), ('cireșe', 15)])

# Parcurgerea perechilor cheie:valoare

print('\n')
for x,y in produse.items():
    print(x,y)

# OBS.: Avem situații în care dorim să folosim doi iteratori pentru o buclă FOR.
# În cazuri de genul, iteratorii vor parcurge treptat elementele și vor lua
# valori pe rând (x = 'mere', y = 5, x = 'pere', y = 7, etc...).
````
