# Capitol 1: Operatori condiționali

Operatorii condiționali sunt folosiți pentru a compara valori și a verifica condiții. Aceștia returnează valori de tip **True** (Adevărat) sau **False** (Fals) și sunt esențiali pentru controlul fluxului unui program.

````python
import pandas as pd

# Definirea datelor pentru tabel
data = {
    'Operator': ['==', '!=', '>', '<', '>=', '<='],
    'Descriere': ['Egalitate', 'Diferit', 'Mai mare decât', 'Mai mic decât', 'Mai mare sau egal', 'Mai mic sau egal'],
    'Exemplu': ['5 == 5 → True', '5 != 4 → True', '7 > 3 → True', '2 < 5 → True', '4 >= 4 → True', '3 <= 4 → True']
}

# Crearea DataFrame-ului
df = pd.DataFrame(data)

# Afișarea tabelului
print(df)

# OBS.: În cazul prezentat, operatorul „==" este diferit de „=" și nu mai are
# funcția de OPERATOR DE ATRIBUIRE, ci în acest caz este OPERATOR DE EGALITATE.
````

````python
# Exemplu 1:

x = 10
y = 20

print("x este egal cu y:", x == y)  # False
print("x este mai mic decât y:", x < y)  # True

# OBS.: Putem identifica faptul că operațiile care folosesc operatori
# condiționali întorc o valoare BOOLEANĂ (valoare de adevăr).
````

Instrucțiunile if, elif și else ne permit să luăm decizii în funcție de anumite condiții.

# Capitol 3: Bucle. De ce le folosim? Cum interacționăm cu ele?

O buclă este o structură care permite repetarea unei secvențe de instrucțiuni de mai multe ori. Bucla continuă să ruleze până când o condiție specificată devine falsă.

## De ce folosim bucle?
*  *Automatizare* – evităm scrierea de cod repetitiv.
*  *Eficiență* – putem itera prin liste, șiruri de caractere sau alte structuri de date.
*  *Flexibilitate* – controlăm de câte ori se execută o anumită porțiune de cod.

## Ce înseamnă DRY?

**DRY** (Don’t Repeat Yourself) este un principiu fundamental în programare care spune că nu trebuie să repetăm codul inutil. În loc să scriem același cod de mai multe ori, ar trebui să-l reutilizăm.

## Cum se aplică principiul DRY în bucle?

Atunci când scriem bucle (for sau while), scopul este să evităm repetarea manuală a aceluiași cod. Dacă observi că o secvență de cod se repetă de mai multe ori, trebuie să folosești o buclă pentru a automatiza procesul.

````python
# Exemplu 1:

print(0)
print(1)
print(2)
print(3)
print(4)

# OBS.: În acest caz, am putea argumenta că este mai rapid să scrim de 5 ori
# comanda print( ). Însă, în alte cazuri, vom avea situații în care vom dori să
# rulăm aceeași bucată de cod de un număr semnificativ crescut de ori, caz în
# care vom apela la o buclă.
````

````python
# Exemplu 2:
for i in range(5): # -> [0, 5) -> 0 1 2 3 [4]
  print(i)

# OBS.: În acest caz, am folosit o buclă FOR să rulăm aceeași bucată de cod de
# 5 ori, dar cu un parametru „i" constant crescător.
````

# Bucla FOR

O buclă **FOR** este o structură de control care permite repetarea unei acțiuni pentru ***un număr cunoscut de iterații***. Cu alte cuvinte, dacă știi de câte ori vrei să repeți o secvență de cod, folosești bucla FOR.



```
for variabila in secventa:
    # Codul care se va executa la fiecare iterație
```
* variabila – La fiecare iterație, această variabilă ia pe rând valorile din secvența specificată.
* secventa – Poate fi o listă, un șir de caractere, o gamă de numere (folosind range()), sau orice obiect iterabil.
* indentația (spațiile la începutul liniilor) – Este necesară pentru a delimita codul care aparține buclei.





````python
for i in range(5): # EXCLUSIVA -> 0 1 2 3 4
    print(f"Valoarea lui i este {i}")
print(i)
# OBS.: Pentru a itera prin mai multe valori, folosim o funcție numită range( )
# În mod standard, această funcție încep de la 0, merge din 1 în 1 și se oprește
# cu 1 înainte de limita impusă.

# În acest caz, începe de la 0, se oprește la 4 și trece prin toate valorile
# din mulțimea {0, 1, 2, 3, 4}.

# Însă, putem manipula acest lucru.
````

## Funcția RANGE
Funcția range() este o funcție încorporată în Python care generează o secvență de numere. Această secvență poate fi folosită în bucle for, pentru a controla câte iterații se fac și care sunt valorile pe care le ia variabila de control. Numerele generate de range() nu sunt salvate într-o listă, ci sunt generate “la cerere” (lazy evaluation), ceea ce o face foarte eficientă.

Funcția range( ) are trei variante principale:
```
range(stop)
range(start, stop)
range(start, stop, pas)
```
În funcție de câți parametri îi dăm, înțelege la care din variante ne referim.

````python
# Exemplu 1 -- Cazul 1

a = 5

for i in range(6): # [0,6) -> 0 1 2 3 4 5
  a += 1  # a = a + 1
  print(a)
  print(f"i = {i}")
  print("__________")

# OBS.: Am folosit un operator atributiv de adunare, care este același lucru cu
# expresia „a = a + 1".
````

````python
# Exemplu 2 -- Cazul 2
a = 5

for i in range(2, 6): # 2 3 4 5
  a = a + 1
  print(a)

# OBS.: Vedem cum putem folosi valorile pe care le ia i în interiorul buclei
# FOR. Această declarare a lui i se spune că are un „scope local", adică o
# putem folosi strict în interiorul FOR și nu există în afara acestuia.
# De obicei, iteratorii sunt denumiți de la litera i încolo (i, j, k ...)
````

````python
# Exemplu 3 -- Cazul 3

a = 5

for i in range(2, 6, 2):  # 2 4
  a = a + i
  print(a)

# OBS.: Aici, am sărit din 2 în 2. Observăm cum nu am luat în considerare și
# cifra 6 ca și valoare a lui i, deoarece este limita maximă a intervalului.
# Deci, singurele valori folosite sunt i = 2 și i = 4.
````

````python
# Exemplu 4 -- Caz special

for i in range(6, 2, -2): # (2, 6]
  print(i)

# OBS.: Putem adăuga un număr negativ și putem schimba ordinea capetelor
# intervalului pentru a face un range( ) care merge descrescător.
````

# Bucla WHILE

O buclă while este o structură de control care execută repetitiv un bloc de cod atâta timp cât o condiție specificată este adevărată (True). Spre deosebire de bucla for, unde știm de la început de câte ori se va executa, bucla while se bazează pe o condiție logică și nu știm neapărat câte iterații vor avea loc.

Sintaxa:
```
while conditie:
    # Codul care se va executa atâta timp cât condiția este True
    # Fiți foarte atenți la identare.
```

## Componente:
	1.	Condiția – Dacă această condiție este True, bucla continuă să ruleze. Când devine False, bucla se oprește.
	2.	Instrucțiunile din buclă – Codul care se execută la fiecare iterație.
	3.	Modificarea condiției – Este important să actualizăm variabilele din condiție, altfel bucla poate deveni infinită.



````python
# Exemplu 1
i = 1

while i <= 5:  # Condiția este True până când i > 5
    print(f"Valoarea lui i este: {i}")
    i += 1  # Creștem valoarea lui i     i = i + 1

# OBS.: Observăm cum putem să folosim operatori condiționali pentru a structura
# condiția de oprire a WHILE.
````

````python
#Exemplu 2
while True:  # Această condiție este mereu True
    print("Aceasta este o buclă infinită!")

# Folosiți butonul din stânga pentru a opri generarea!
````

````python
# __EXERCIȚIU__ Se da un numar i = 0 si un numar a = 5. Cat timp a este mai
# mic decat 100, adaugati valoarea lui i la a si cresteti valoarea lui i cu 1
# in fiecare ciclu. Afisati rezultatul.

i = 0
a = 5

while a < 100:
  a = a + i # a += i
  i = i + 1 # i += 1

  print(f"Ciclul {i}: a = {a}")

print()
print(f"Rezultatul final: a = {a}")
````

# Controlul buclei. BREAK. CONTINUE. PASS

1. BREAK -- Un mod sigur de a opri bucla cu totul
Când se întâlnește break, bucla se oprește imediat, chiar dacă condiția nu a devenit False.

````python
i = 1
while True:
    print(f"i = {i}")
    if i == 5:
        print("Oprirea buclei cu BREAK.")
        break
    i += 1



# OBS.: Am folosit aceeași condiție „infinită" ca mai sus pentru WHILE, însă
# de această dată am forțat oprirea buclei când am ajuns la rezultatul dorit.
````

2. CONTINUE -- Un mod de a sări peste iterația actuală a buclei fără a ieși de tot din ea

````python
i = 0
# i = 0 -> +1 -> i = 1 -> "Numar impar: 1"
# i = 1 -> +1 -> i = 2 -> continue (am sarit peste print)
# i = 2 -> +1 -> i = 3 ...


while i < 10:
    i += 1

    if i % 2 == 0:
        continue
    print(f"Număr impar: {i}")

# OBS.: Am evitat cu ușurință toate numerele pare folosind CONTINUE. Într-un
# caz similar cu acesta ar fi mai bine să lucrăm invers, adică să rulăm doar
# dacă o condiție se întâmplă (adică am modifica IF-ul cu „i % 2 == 1"), pentru
# că această situație scade înțelegerea codului de către un alt dezvoltator.
````

3. PASS -- O instrucțiune specifică Python care permite includerea de bucle sau condiții care nu au conținut. Când se întâlnește PASS, se execută codul fără a întoarce o eroare, deoarece sare automat din bucla în care îl găsește.
* Se folosește mai mult pentru a crea structuri de algoritm, pentru cazurile în care nu știm încă ce să adăugăm în interiorul algoritmului.

````python
i = 0
while i < 5:
    if i == 2:
        pass
    print(f"Valoarea lui i este {i}")
    i += 1

for j in range(1,100):
  pass


# !! BUCLA INFINITA
# j = 0
# while j < 10:
#   pass


# OBS.: Codul a rulat cum trebuie, chiar dacă nu avem nimic în interiorul
# condiției IF. Este de evitat o astfel de scriere deoarece poate fi
# uitată în interiorul algoritmului.
# Se folosește la interviuri de către angajator pentru a structura forma dorită
# a codului.
````

````python
# __EXERCIȚIU__ Fie i = 0. Cresteti valoarea lui i cu 2 pana intalneste un
# numar divizibil cu 5. Cand a fost gasit, iesiti din bucla si afisati
# rezultatul.

i = 2

while i % 5 != 0:
  i += 2

print(i)


i = 0

while True:
  i += 2
  if i % 5 == 0:
    break

print(i)
````
