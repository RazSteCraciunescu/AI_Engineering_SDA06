#Parcurgerea unui Șir de Caractere (String) folosind Bucla FOR
##Ce înseamnă parcurgerea unui string?

Un string în Python este o secvență de caractere. Fiecare caracter dintr-un string are o poziție (index) și poate fi accesat individual. Prin utilizarea buclei for, putem parcurge caracterele unu câte unu și aplica diverse operații asupra lor.

##De ce folosim bucla for pentru a parcurge un string?
	1.	Accesarea fiecărui caracter în mod individual.
	2.	Aplicarea de acțiuni asupra fiecărui caracter (de exemplu, afișare, verificare, comparare etc.).
	3.	Evita repetarea inutilă de cod (principiul DRY).
	4.	Ușor de înțeles și folosit.


  ### Sintaxa
```
for caracter in string:
    # Acțiune asupra fiecărui caracter
```

* string – Este șirul de caractere pe care dorim să-l parcurgem.
*	caracter – La fiecare iterație, această variabilă preia câte un caracter din string.
* Corpul buclei – Aici se află codul care se execută pentru fiecare caracter.

````python
# EXEMPLU 1

text = "Python !"

# 1. caracter = "P" -> print ...
# 2. caracter = "y" -> print ...
# ...
# 7. caracter = " " -> print ...
# 8. caracter = "!" -> print ...
# STOP

for caracter in text:
    print(f"Caracter: {caracter}")
    caracter = " "

print(text)

# OBS.: Variabila introdusă ca și iterator ia pe rând valoarea fiecărui
# CARACTER din textul dat. De aceea, putem vedea că atât caracterul " ",
# cât și "!" sunt luate în considerare.
````

````python
# EXEMPLU 2

text = "Python !"

# range(len(text)) -> range(8)
# range(8) -> [0, 8) -> 0 1 2 3 4 5 6 7

# i = 0 -> text[0] = "P"
# i = 1 -> text[1] = "y"
# ...
# i = 7 -> text[7] = "!"

for i in range(len(text)):
    print(f"Caracter: {text[i]}") # text[0] text[1] text[2] text[3] ...

# OBS.: Putem folosi funcția range( ) pentru a parcurge toate POZIȚIILE unui
# STRING. Numerotarea pozițiilor începe de la 0 până la (lungime-1), ca în
# schema de mai jos:
# P y t h o n   !
# 0 1 2 3 4 5 6 7

# Am folosit funcția len( ) pentru a determina lungimea string-ului. Rezultatul
# funcției len(text) va fi 8, însă ne amintim că funcția range( ) merge strict
# până la (limita_maxima - 1), care în acest caz ne avantajează pentru că noi
# dorim să parcurgem toate pozițiile șirului.
````

````python
# __EXERCIȚIU__ Se citește un cuvânt w DE LA TASTATURĂ. Să se parcurgă literele
# acestui cuvânt și, dacă se găsește litera „a" sau „A", să se afișeze pe ecran
# „Am găsit!" și să se oprească parcurgerea. Dacă s-a parcurs tot cuvântul și
# nu s-a găsit, să se afișeze „Cuvântul [...] nu conține litera a/A!".

litera_a_gasita = False

# HINT: Putem folosi operatorul de egalitate „==" în lucrul cu string-uri.

# copac

w = input()

for litera in w:
  if litera == "a" or litera == "A":
    print(f"Am gasit {litera}!")
    litera_a_gasita = True
    break
  else:
    print(f"Litera {litera} nu este conforma!")

if litera_a_gasita == False:
  print(f"Cuvântul {w} nu conține litera a/A!")
````

# Ce este ASCII și cum aflăm valoarea ASCII a unui caracter?

ASCII (American Standard Code for Information Interchange) este un sistem de codificare folosit pentru a reprezenta caracterele (litere, cifre, simboluri speciale) ca valori numerice. Fiecare caracter are un cod numeric unic, ceea ce permite calculatoarelor să stocheze și să manipuleze text.

##Cum funcționează ASCII?
	•	Codurile ASCII sunt numere de la 0 la 127.
	•	Caracterele ASCII standard includ literele mari și mici, cifrele, simbolurile și caractere de control (cum ar fi “enter” sau “backspace”).
	•	De exemplu:
	•	Litera A are codul 65.
	•	Litera a are codul 97.
	•	Simbolul @ are codul 64.

##Cum aflăm valoarea ASCII a unui caracter în Python?

Pentru a obține valoarea ASCII a unui caracter, folosim funcția ord(caracter). Această funcție returnează codul ASCII corespunzător caracterului.

Sintaxă
```
ord(caracter)
```
* caracter – Un caracter individual (de exemplu, ‘A’, ‘a’, ‘1’, ‘$’ etc.).
* Returnează – Valoarea ASCII (un număr întreg) a acelui caracter.


````python
# EXEMPLU 1

caracter = 'A'
valoare_ascii = ord(caracter)
print(f"Valoarea ASCII a caracterului '{caracter}' este {valoare_ascii}")
````

````python
# EXEMPLU 2

valoare_ascii = 65
caracter = chr(valoare_ascii)
print(f"Caracterul pentru valoarea ASCII {valoare_ascii} este '{caracter}'")

# OBS.: Am folosit funcția chr( ) pentru a converti o valoare (număr întreg)
# în echivalentul în ASCII al acelei valori.
````

````python
# __EXERCIȚIU__ Se citește un număr natural n > 0 de la tastatură. Pe urmă,
# utilizatorul este rugat de n ori să introducă de la tastatură câte o literă.
# După fiecare input de la utilizator, se va afișa:
# „Litera următoare în alfabet este: [...]"
# Dacă s-a introdus „z" răspunsul va fi „a", iar dacă s-a introdus „Z"
# răspunsul este „A".

# n = 3

# 1. litera = "a" -> "b"
# 2. litera = "C" -> "D"
# 3. litera = "Z" -> "A"

n = int(input("Introdu un numar de litere: "))

for i in range(n):
  litera = input(f"Introdu litera {i+1}:")

  if litera == "z":
    print(f"Litera următoare în alfabet este: a")

  elif litera == "Z":
    print(f"Litera următoare în alfabet este: A")

  else:
    cod_ascii = ord(litera)
    cod_ascii = cod_ascii + 1
    litera = chr(cod_ascii)
    print(f"Litera următoare în alfabet este: {litera}")



# HINT: O variantă de a executa acest exercițiu este de a converti răspunsul
# primit de la utilizator în cod ASCII, să creștem codul ASCII cu o unitate și
# să convertim numărul obținut înapoi în caracter.
````
